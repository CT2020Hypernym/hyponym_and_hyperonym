### Идея: для каждого слова(гипонима и гиперонима) найти 1 контекст из википедии  в отдельный файл с названием из этого слова, чтобы избежать memory error.

```python
path = "contexts_from_wiki//pos//"
```

```python
words = set()
for i in range(len(positive.child)):
    words.add(positive.child[i])
    words.add(positive.parent[i])
```

```python
for i in words:
    print(i)
```

```python
for i in words:
    with open('wikipedia-ru-2018.txt', 'r', encoding = 'utf-8') as g:
        with open(path + i + '.txt', 'w', encoding = 'utf-8') as f:
            t = ''
            c = 0
            while(i not in t):
                c += 1
                t = g.readline()
                if c > 20908028:
                    t =''
                    break
            f.write(t)
```

```python
import time
start_time = time.time()
print(sum(1 for line in open('wikipedia-ru-2018.txt', 'r', encoding = 'utf-8')))
print("--- %s seconds ---" % (time.time() - start_time))
```
