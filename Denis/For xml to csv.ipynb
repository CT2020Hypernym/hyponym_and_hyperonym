```python
import pandas as pd
import itertools
from bs4 import BeautifulSoup as b
import os
```

```python
with open('synset_relations.A.xml', "r", encoding='utf-8') as f: # opening xml file
    content = f.read()
soup = b(content, "lxml")
```

```python
relations = soup.find_all('relation')
```

```python
relations[0].attrs
```

```python
for_csv = pd.DataFrame()
for i in range(len(relations)):
    if i%1000 == 0:
        print(i)
    for_csv = for_csv.append(relations[i].attrs, ignore_index=True)
```

```python
for_csv.head()
```

```python
#for_csv.to_csv(os.getcwd(), index= None, compression=compression_opts)
for_csv.to_csv(os.getcwd()+"\\synset_relations.A.csv" , encoding='utf-8', header=True, index=False)
```
