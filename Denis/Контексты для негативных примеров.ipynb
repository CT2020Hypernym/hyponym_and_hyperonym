```python
import pandas as pd
```

```python
df = pd.read_csv('neg.csv', encoding='utf-8')
```

```python
words = set()
```

```python
for i in range(len(df.child)):
    words.add(df.child[i])
    words.add(df.parent[i])
```

```python
path = "contexts_from_wiki//neg//"
```

```python
for i in words:
    with open('wikipedia-ru-2018.txt', 'r', encoding = 'utf-8') as g:
        with open(path + i + '.txt', 'w', encoding = 'utf-8') as f:
            t = ''
            c = 0
            while(i not in t):
                c += 1
                t = g.readline()
                if c > 20908028:
                    t =''
                    break
            f.write(t)
```
